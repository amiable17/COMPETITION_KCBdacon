```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data_Loading

```python
df = pd.read_csv("credit_card_data.csv", encoding="UTF-8")
```

```python
df.head()
```

```python
df['org_total_loan'] = df['monthly_bk_loan'] + df['monthly_cd_loan'] + df['monthly_installments_loan'] + df['monthly_insurance_loan'] +df['monthly_sbk_loan']
df['org_loan_amt_error'] = df['monthly_loan'] - df['org_total_loan']
```

```python
df['by_total_loan'] = df['credit_loan'] + df['mortgage_loan'] 
df['by_loan_amt_error'] = df['monthly_loan'] - df['by_total_loan']
```

```python
def divs_data_set(x):
    data_dvis_cd = None
    
    if(x[0] == "L"):
        data_dvis_cd = "region_age"
    elif (x[0] == "Y"):
        data_dvis_cd = "age_sex"
    else:
        data_dvis_cd = "Error"
    
    return data_dvis_cd

df['data_dvis_cd'] = df['pop_cd'].map(lambda x : divs_data_set(x))
```

```python
region_age_df = df[df['data_dvis_cd']=="region_age"]
age_sex_df = df[df['data_dvis_cd']=="age_sex"]
```
